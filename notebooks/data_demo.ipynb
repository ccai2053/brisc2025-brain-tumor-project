# A Demo of the BRISC 2025 Data Loader

This notebook demonstrates loading brain MRI images from the BRISC 2025 dataset using a custom PyTorch DataLoader.

Inputs:
- Brain MRI images

Targets:
- Tumor classification labels

```python
import sys
sys.path.append("../src")
from brisc_project.brisc import get_data_loader
```

```python
train_dir = "../data/sample/train"

example_loader, classes = get_data_loader(
    train_dir,
    batch_size=5
)

images, labels = next(iter(example_loader))

print("Image batch shape:")
print(images.shape)

print("\nLabel batch shape:")
print(labels.shape)

print("\nClasses:")
print(classes)

print("\nLabels:")
print(labels)
```

```python
from matplotlib import pyplot as plt
import numpy as np

plot_images = images.numpy().transpose(0, 2, 3, 1)

plt.figure(figsize=(15, 5))
plt.imshow(np.hstack(plot_images))

for index, label in enumerate(labels):
    plt.text(
        index * 224 + 50,
        20,
        classes[label.item()],
        color="red",
        fontsize=10,
        fontweight="bold"
    )

plt.axis("off")
plt.savefig("../assets/sample_batch.png")
plt.show()
```

The images above show a batch of MRI scans loaded using the custom PyTorch DataLoader, along with their corresponding tumor classification labels.
